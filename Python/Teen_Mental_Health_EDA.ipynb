Summary: The data set focused on the impact social media has on teen mental health. Its markers included age, daily social media hours, sleep hours, screen time before sleep, academic performance, physical activity, stress level, anxiety level, addictional level, and depression label. The analysis revealed the relatonship between these factors is generally weak as there is no significant direct impact on teen mental health facilitated by social media or related behaviors.

Recommendation: Mental health is affected by a multitude of markers and though social media may be one of them, it would be better to it alongside other influencing factors in order to capture a more complete picture. Additionally, it would ensure mental health is understood as a complex and layered issue that requires comprehensive analizying of the factors that impact it.
